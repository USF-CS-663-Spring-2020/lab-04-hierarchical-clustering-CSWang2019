```python
import pandas as pd

df = pd.read_csv('USCensus1990-2000rows.data.txt')
# df = df.loc[:10]
df
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(df, 'ward')
```

```python
# Draw the dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(20,20))
dendrogram(Z,leaf_rotation=90, leaf_font_size=8) # let label ratate by 90 degree
# plt.show()#decrete
plt.axhline(y=15000, c='grey', lw=1, linestyle='dashed') 
# we can see from the plot, when y = 1500, it can be splited into three clusters
```

```python
# dendrogram( ) # -> you will have 4 leaf at the bottom of the plot
plt.figure(figsize=(20,20))
dendrogram(Z, truncate_mode = 'lastp', p=10, labels=df.index)
# we also can drow dendrogram by setting leaves number
```

```python
import pandas as pd
df = pd.read_csv('NIPS_1987-2015.csv')
df
```

```python
df = df.drop(columns = 'Unnamed: 0') #drop the line with strings
df
# df.drop(columns=5000)
#drop(columns=['B', 'C'])
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
Z = linkage(df, 'ward') 
# transfer dataframe into hierarchy data  
```

```python
# Draw the dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(20,200))
dendrogram(Z, orientation='right')
plt.show()
# rotate the dendrogram for visibility 
```
