```python
import pandas as pd

df = pd.read_csv('USCensus1990-2000rows.data.txt')
# df = df.loc[:10]
df
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(df, 'ward')
```

```python
# Draw the dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(20,10))
dendrogram(Z)
plt.show()#decrete
```

```python
import pandas as pd

df = pd.read_csv('NIPS_1987-2015.csv')
df
```
